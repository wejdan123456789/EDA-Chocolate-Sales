# Assignment 1: Exploratory Data Analysis (EDA)

EDA is the first and most important step in any Machine Learning project.
Before building models, we must understand:

- What does the data represent?
- Are there missing values?
- Are there outliers?
- What patterns exist?
- Which variables influence others?

If we do not understand the data, we cannot build a good model.

### Why EDA is Important

In real-world machine learning projects:

    1- 70–80% of the time is spent on understanding and cleaning data
    2- Only 20–30% is spent building models

**EDA helps us:**

- Detect errors
- Identify trends
- Discover relationships
- Make business decisions

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset
df = pd.read_csv("Chocolate_Sales.csv")
# Display first 5 rows. If you want to display all rows, use df.head(-1)
df.head(-1)
```

### Check Missing Values

```python
# Check all values
print(df.isna())
```

```python
print(df.isna().sum())
```

```python
df.info()
df.describe()
df.columns
```

### Check duplicate rows

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

### No. of rows and columns

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

### Data type of columns

```python
# viewing the data types of columns
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Amount'] = df['Amount'].replace(r'[\$,]', '', regex=True)
df['Amount'] = pd.to_numeric(df['Amount'])

df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

### Data Cleaning

```python

df = df.dropna()

if df['Amount'].dtype == 'object':
    df['Amount'] = df['Amount'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)


df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

print(df.isnull().sum())
df.info()
```

Cleaning was needed because the 'Amount' column was a string with '$' symbols, and there were missing values in the 'Country' column."

```python
df.head()
df.info()
```

### data analysis

```python
top_products = df.groupby('Product')['Amount'].sum().sort_values(ascending=False).head(5)
print(top_products)

country_sales = df.groupby('Country')['Amount'].sum().sort_values(ascending=False)
print(country_sales)
```

The analysis reveals that India is the top-performing market, while Mint Chip Choco is the most popular product. These results suggest focusing marketing efforts on high-demand regions and top-selling products to maximize total revenue.

### Descriptive summary Statistics

### Pie Chart (Sales Share by Product)

```python
plt.figure(figsize=(8,8))
product_revenue = df.groupby('Product')['Amount'].sum()
product_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Sales Percentage by Product')
plt.ylabel('')
plt.show()
```

Goal: To determine the market share and contribution of each product to the total revenue.

Insight: Mint Chip Choco holds the largest share of sales, making it the most successful product in the portfolio.

### Horizontal Bar Chart (Top 10 Sales Persons)

```python
plt.figure(figsize=(10,8))
df.groupby('Sales Person')['Amount'].sum().nlargest(10).sort_values().plot(kind='barh', color='green')
plt.title('Top 10 Sales Persons Performance')
plt.show()
```

Goal: To evaluate and rank the performance of individual sales representatives.

Insight: A small group of top performers contributes to a significant portion of the total sales, highlighting their efficiency.

### Line Plot (Monthly Sales Trend)

```python
plt.figure(figsize=(12,6))
df.set_index('Date')['Amount'].resample('ME').sum().plot(marker='o', color='red')
plt.title('Monthly Sales Trend')
plt.grid(True)
plt.show()
```

Goal: To track sales performance over time and identify seasonal patterns or trends.

Insight: The chart shows fluctuations in monthly revenue, which helps in planning inventory and marketing campaigns.

### Distribution of Sales Amount (Histogram)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['Amount'], bins=20, kde=True, color='blue')

plt.title('Distribution of Sales Amount')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

Goal: To visualize the frequency and spread of individual transaction values.

Insight: Most sales transactions fall within a specific mid-range, indicating a consistent pricing strategy.

## Bivariate Analysis

### Total Sales by Country (Bar Chart)

```python
plt.figure(figsize=(10,6))
country_sales = df.groupby('Country')['Amount'].sum().sort_values(ascending=False)

sns.barplot(x=country_sales.index, y=country_sales.values, hue=country_sales.index, palette='magma', legend=False)

plt.title('Total Sales by Country')
plt.xticks(rotation=45)
plt.show()
```

Goal: To compare the total revenue generated across different geographical regions.

Insight: India is the leading market with the highest sales volume, while other regions show potential for growth.

### Scatter Plot (Relationship: Boxes Shipped vs. Amount)

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Boxes Shipped', y='Amount', hue='Country', alpha=0.7)
plt.title('Relationship: Boxes Shipped vs. Sales Amount')
plt.show()
```

Goal: To examine the correlation between the quantity of units sold and the total revenue.

Insight: There is a strong positive correlation, meaning that higher shipping volumes directly lead to increased profits.

### Box Plot (Sales Amount Spread by Country)

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Country', y='Amount')
plt.title('Sales Amount Distribution per Country')
plt.xticks(rotation=45)
plt.show()
```

Goal: To detect outliers and understand the distribution of sale prices within each country.

Insight: Some countries have a wider range of transaction values, suggesting diverse customer purchasing power in those markets.
